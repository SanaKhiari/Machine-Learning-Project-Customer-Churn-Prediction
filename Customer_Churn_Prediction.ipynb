# <span style="color: #2E86C1; font-weight: bold;">Customer Churn</span>


Customer churn is a **significant challenge** faced by telecom companies. Churn occurs when customers stop using the company's services and switch to competitors.

## <span style="color: #D35400; font-weight: bold;">The Causes of Customer Churn in Telecommunications Are Varied:</span>

- **Pricing**
- **Dissatisfaction with service quality**
- **Poor Customer Experience**
- **Better Competitor Offerings**

- **Marketing Manager**
- **General Management**
- **Product Manager**


## <span style="color: #16A085; font-weight: bold;">A Deeper Understanding of Churn Can Help Stakeholders in Several Ways:</span>

- Identifies and addresses features that fall short of customer expectations.
- Ensures proactive measures are taken to prevent dissatisfaction from spreading to other customers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn-bigml-20.csv')
df1 = pd.read_csv('churn-bigml-80.csv')
df_combined = pd.concat([df, df1], ignore_index=True)
df_combined.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Identify numerical columns
numerical_cols = df_combined.select_dtypes(include=[np.number]).columns.tolist()

# Loop over each numerical column and plot separately
for col in numerical_cols:
    plt.figure(figsize=(10, 6))

    # Create histogram with KDE
    sns.histplot(df_combined[col], kde=True, color="#FFC8DD", edgecolor="black", linewidth=1.5, bins=30)

    # Add title and axis labels with improved formatting
    plt.title(f'Distribution of {col}', fontsize=18, fontweight='bold', color='black')
    plt.xlabel(col, fontsize=14, fontweight='bold', color='black')
    plt.ylabel('Frequency', fontsize=14, fontweight='bold', color='black')

    # Customize grid and background
    plt.grid(axis='y', linestyle='--', linewidth=0.8, alpha=0.8)

    # Add annotations for peak values (max density point in KDE)
    if df_combined[col].notnull().any():  # Ensure the column is not empty
        kde_ax = sns.kdeplot(df_combined[col], color="lightgreen", linewidth=2)
        max_density = kde_ax.lines[0].get_data()[1].max()  # Find max y-value (density)
        max_x = kde_ax.lines[0].get_data()[0][kde_ax.lines[0].get_data()[1].argmax()]  # Corresponding x-value
        plt.annotate(f"Peak: {max_x:.2f}", xy=(max_x, max_density), xytext=(max_x + 0.5, max_density + 0.02),
                     arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12, color="black", fontweight="bold")

    # Adjust layout to avoid clipping
    plt.tight_layout()
    plt.show()
```

### Label Encoding
Label Encoding is a technique that converts categorical values into numerical values by assigning each category a unique integer.

```python
d = {"No": 0, "Yes": 1}
df_combined["International plan"] = df_combined["International plan"].map(d)
df_combined = df_combined.replace({"Voice mail plan": d})

# Churn = 0: Customers who did not churn.
# Churn = 1: Customers who churned
df_combined["Churn"] = df_combined["Churn"].astype("int64")
df_combined.head()
```

### Frequency Encoding
A technique where categorical values are replaced by the frequency of their occurrence in the dataset.

```python
# Frequency Encoding : remplacer les valeurs dans la colonne 'State'
frequency_encoding = df_combined["State"].value_counts()    # Calcul des fr√©quences
df_combined["State"] = df_combined["State"].map(frequency_encoding)  # Remplacer les cat√©gories par leurs fr√©quences

# V√©rification
print("Colonne 'State' apr√®s Frequency Encoding :")
print(df_combined.head())
```

```python
import joblib

# Create the frequency encoding dictionary
state_freq_mapping = df_combined["State"].value_counts().to_dict()

# Save it
joblib.dump(state_freq_mapping, "state_freq_mapping.pkl")
```

 Standardize all numerical features in the dataset (excluding the target variable "Churn") using StandardScaler from sklearn.preprocessing, ensuring that each feature has a mean of 0 and a standard deviation of 1 for better model performance. üöÄ

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()


numeric_columns = df_combined.select_dtypes(include=['float64', 'int64']).columns

# Exclure la variable cible 'Churn' des colonnes √† normaliser
numeric_columns = numeric_columns.difference(['Churn'])

# Appliquer la standardisation sur les colonnes num√©riques (hors Churn)
df_combined[numeric_columns] = scaler.fit_transform(df_combined[numeric_columns])
```

### data splitting and balancing the data


```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

# features
X = df_combined.drop(columns=["Churn"])
#target
Y = df_combined["Churn"]



# ‚úÖ Step 1: Apply Random Oversampling to balance minority class
ros = RandomOverSampler(sampling_strategy='minority', random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, Y)

# ‚úÖ Step 2: Apply SMOTEENN for synthetic oversampling + noise reduction
smoteenn = SMOTEENN(random_state=42)
X_balanced, y_balanced = smoteenn.fit_resample(X_resampled, y_resampled)

# üîç Show class distribution before and after balancing
print("Before balancing:", Counter(Y))
print("After balancing:", Counter(y_balanced))

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)


print("train : " ,X_train.shape, Y_train.shape)
print("test : " , X_test.shape, Y_test.shape)

```

## Logistic Regression

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ‚úÖ Train the Logistic Regression model
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, Y_train)

# ‚úÖ Make predictions
y_pred_log = logreg.predict(X_test)

# ‚úÖ Evaluate the model
accuracy_log = accuracy_score(Y_test, y_pred_log)
class_report_log = classification_report(Y_test, y_pred_log)
conf_matrix_log = confusion_matrix(Y_test, y_pred_log)

print("\nüîπ Accuracy Score:", accuracy_log)
print("\nüîπ Classification Report:\n", class_report_log)
print("\nüîπ Confusion Matrix:\n")

# ‚úÖ Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_log, annot=True, fmt="d", cmap="YlGnBu", xticklabels=[0, 1], yticklabels=[0, 1])

# ‚úÖ Labels and Titles
plt.title("Confusion Matrix: Logistic Regression", fontsize=14, fontweight='bold')
plt.xlabel("Predicted Label", fontsize=12, fontweight='bold')
plt.ylabel("True Label", fontsize=12, fontweight='bold')

# ‚úÖ Show the Plot
plt.show()
```

# Decision tree


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

dt_classifier = DecisionTreeClassifier(criterion="gini", max_depth=5, random_state=42)
dt_classifier.fit(X_train, Y_train)

# ‚úÖ Step 5: Make Predictions
Y_pred = dt_classifier.predict(X_test)

# ‚úÖ Evaluate Model
accuracy = accuracy_score(Y_test, Y_pred)
class_report = classification_report(Y_test, Y_pred)
conf_matrix = confusion_matrix(Y_test, Y_pred)

print("\nüîπ Accuracy Score:", accuracy)
print("\nüîπ Classification Report:\n", class_report)
print("\nüîπ Confusion Matrix:\n")

# ‚úÖ Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Oranges", xticklabels=[0, 1], yticklabels=[0, 1])

# ‚úÖ Labels & Titles
plt.title("Confusion Matrix: Decision Tree Classifier", fontsize=14, fontweight='bold')
plt.xlabel("Predicted Label", fontsize=12, fontweight='bold')
plt.ylabel("True Label", fontsize=12, fontweight='bold')

# ‚úÖ Show the Plot
plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ‚úÖ Train the Random Forest model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, Y_train)

# ‚úÖ Make predictions
y_pred_rf = rf_classifier.predict(X_test)

# ‚úÖ Evaluate the model
accuracy_rf = accuracy_score(Y_test, y_pred_rf)
class_report_rf = classification_report(Y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(Y_test, y_pred_rf)

# Print evaluation metrics
print("\nüîπ Accuracy Score:", accuracy_rf)
print("\nüîπ Classification Report:\n", class_report_rf)
print("\nüîπ Confusion Matrix:\n", conf_matrix_rf)

# ‚úÖ Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="YlGnBu", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])

# ‚úÖ Customize plot
plt.title("Confusion Matrix: Random Forest", fontsize=14, fontweight='bold')
plt.xlabel("Predicted Label", fontsize=12, fontweight='bold')
plt.ylabel("True Label", fontsize=12, fontweight='bold')

# ‚úÖ Show the Plot
plt.show()
```

## XGBoost

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ‚úÖ Train the XGBoost model
xgb_classifier = xgb.XGBClassifier(random_state=42)
xgb_classifier.fit(X_train, Y_train)

# ‚úÖ Make predictions
y_pred_xgb = xgb_classifier.predict(X_test)

# ‚úÖ Evaluate the model
accuracy_xgb = accuracy_score(Y_test, y_pred_xgb)
class_report_xgb = classification_report(Y_test, y_pred_xgb)
conf_matrix_xgb = confusion_matrix(Y_test, y_pred_xgb)

# Print evaluation metrics
print("\nüîπ Accuracy Score:", accuracy_xgb)
print("\nüîπ Classification Report:\n", class_report_xgb)
print("\nüîπ Confusion Matrix:\n", conf_matrix_xgb)

# ‚úÖ Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_xgb, annot=True, fmt="d", cmap="YlGnBu", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])

# ‚úÖ Customize plot
plt.title("Confusion Matrix: XGBoost", fontsize=14, fontweight='bold')
plt.xlabel("Predicted Label", fontsize=12, fontweight='bold')
plt.ylabel("True Label", fontsize=12, fontweight='bold')

# ‚úÖ Show the Plot
plt.show()
```

## Compare Models Using ROC Curve

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Pre-trained models
models = {
    "Logistic Regression": logreg,  # Assuming logreg is already trained
    "Random Forest": rf_classifier,  # Assuming rf_classifier is already trained
    "XGBoost": xgb_classifier,  # Assuming xgb_classifier is already trained
    "Decision Tree": dt_classifier  # Assuming dt_classifier is already trained
}

# Plot ROC curve for each model
plt.figure(figsize=(10, 8))

for model_name, model in models.items():
    # Predict probabilities for the positive class
    y_prob = model.predict_proba(X_test)[:, 1]

    # Compute ROC curve
    fpr, tpr, _ = roc_curve(Y_test, y_prob)

    # Compute AUC (Area Under the Curve)
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve for this model
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Plotting the diagonal line (random classifier)
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier (AUC = 0.50)')

# Customize plot
plt.title("ROC Curve Comparison", fontsize=14, fontweight='bold')
plt.xlabel("False Positive Rate", fontsize=12, fontweight='bold')
plt.ylabel("True Positive Rate", fontsize=12, fontweight='bold')
plt.legend(loc="lower right")

# Show plot
plt.show()
```
